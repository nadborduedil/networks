```python
def outcome(left, right, all_coins):
    L = sum(all_coins[i] for i in left)
    R = sum(all_coins[i] for i in right)
    if L > R:
        return 1
    elif L < R:
        return -1
    else:
        return 0
    
def full_outcome(weights, all_coins):
    return tuple(outcome(l, r, all_coins) for l, r in weights)

def full_split(weights, possibilities):
    outcome_to_pos = {}
    for pos in possibilities:
        out = full_outcome(weights, pos)
        if out not in outcome_to_pos:
            outcome_to_pos[out] = []
        outcome_to_pos[out].append(pos_str(pos))
    return outcome_to_pos

def pos_str(possibility):
    for i, w in enumerate(possibility):
        if w > 0:
            return "%s+" % i
        elif w < 0:
            return "%s-" % i
```
